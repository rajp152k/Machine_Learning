```python
import numpy as np
```

```python
points = np.arange(-5,5,0.01)
```

```python
xs,ys = np.meshgrid(points,points)
```

```python
z = np.sqrt(xs**2+ys**2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(z,cmap=plt.cm.gray);plt.colorbar()
plt.title('representing $\sqrt{x^2+y^2}$ for a grid of values')
```
