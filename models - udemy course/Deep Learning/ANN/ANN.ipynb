````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````
import io
dataset = pd.read_csv(io.BytesIO(uploaded['Churn_Modelling.csv']))
````

## Data Preprocessing

````
dataset
````

don't need the customer id, surname and row number in the dependent variable matrix

````
X = dataset.iloc[:,3:-1].values
````

````
Y = dataset.iloc[:,-1].values
````

Preparing matrix of featues(encoding and handling dummy variables)

````
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
````

````
L_encoder_x1 = LabelEncoder()
L_encoder_x2 = LabelEncoder()
````

````
X[:,1] = L_encoder_x1.fit_transform(X[:,1])
X[:,2] = L_encoder_x2.fit_transform(X[:,2])
````

````
X
````

````
OHcoder = OneHotEncoder(categorical_features = [1])
````

````
X = OHcoder.fit_transform(X).toarray()
````

````
pd.DataFrame(X)
````

Dropping the first column of X(dummy var trap)

````
X = X[:,1:]
````

````
pd.DataFrame(X)
````

````
from sklearn.model_selection import train_test_split
````

````
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size= 0.2,random_state=0)
````

````
from sklearn.preprocessing import StandardScaler
````

````
X_scaler = StandardScaler()
X_train = X_scaler.fit_transform(X_train)
X_test = X_scaler.transform(X_test)
````

## Creating the ANN

````
import keras
````

````
from keras.models import Sequential
from keras.layers import Dense
````

## Sequential for initializing the neural network 
## Dense for adding layers to the network

initializing the ANN by defining it as a sequence of layers<br>
creating a classifier

````
classifier = Sequential()
````

input layer has 11 nodes (no. of cols in feature matrix)

for the hidden layers, choosing the activating function to be a rectifier, whereas for the final output node, choosing sigmoid as its activation function to get a probability

For now, the hidden layer is chosen to have 6 nodes, will consider the effect of this when reassess this implementation during studying parameter tuning

````
classifier.add(Dense(units=6,input_dim = 11, activation="relu",kernel_initializer="uniform"))
````

````
classifier.add(Dense(units=6, activation="relu",kernel_initializer="uniform"))
````

no need to add the input dim as we already have inserted the first layer 

adding the output layer now

````
classifier.add(Dense(units=1,activation="sigmoid",kernel_initializer="uniform"))
````

for the final output layer having more than 2 nodes, we use the softmax activation function(sigmoid for multiclassification)

## applying Stochastic gradient descent

````
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics = ["accuracy"] )
````

adam is a sort of a stochastic gradient descent algorithm <br>
binary_crossentropy is used as a loss function for binary classification and multiple_crossentropy for poly-classification

````
classifier.fit(X_train,Y_train,batch_size=10,nb_epoch=100)
````

````
Y_pred = classifier.predict(X_test)
````

````
from sklearn.metrics import confusion_matrix
````

as Y_pred is the result of a sigmoid activation function, it returns probabilities, hence we need to discretize it<br>
could use 
   ```
   Y_pred = (Y_pred>threshold)
   ```
   for an inplace update rather than looping as below

````
Y_pred_discrete = []
for i in Y_pred:
    if i>=0.5:
        Y_pred_discrete.append(1)
    else:
        Y_pred_discrete.append(0)
Y_pred_discrete = np.asarray(Y_pred_discrete)
````

````
Y_pred
````

````
Y_pred_discrete
````

````
cm = confusion_matrix(Y_test,Y_pred_discrete)
````
